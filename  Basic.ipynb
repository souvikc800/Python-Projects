## Independent T-Test<br>

Take 1-D arrays of the attributes for which u want to perform T-Test.<br>

if categorical colm, then get the numerical data from it.



```python
H0 = "charges of smoker and non-smoker are same"

Ha = "charges of smoker and non-smoker are not same"

x =np.array(ins[ins.smoker == 'yes'].charges)

y =np.array(ins[ins.smoker == 'no'].charges)

t ,p =stats.ttest_ind(x,y)

if p < 0.05:
    print(H0,"is false, null hypo rejected, charges are different for smoker and non-smokers")
else:
    print("null is accepted")
```

## Does BMI change with Gender

```python
H0 = "BMI same with gender"

Ha = "BMI changes with gender"

x =np.array(ins[ins.sex == 'male'].bmi)

y =np.array(ins[ins.sex == 'female'].bmi)

t ,p =stats.ttest_ind(x,y)

if p < 0.05:
    print(H0,"is false, null hypo rejected, charges are different for smoker and non-smokers")
else:
    print(H0, " is true , null is accepted")
```

### Chi-square test

```python
Ho="Region has no eefect on smoking habit"

Ha="Region has effect on smoking habits"

crosstab=pd.crosstab(ins.smoker, ins.region)

chi , p , dof , expected = stats.chi2_contingency(crosstab)

if p < 0.05:
    print(Ho,"is rejected")
else:
    print(Ho,"is accepted")
```

```python
Ho="Gender has no eefect on smoking habit"

Ha="Gender has effect on smoking habits"

crosstab=pd.crosstab(ins.sex, ins.smoker)

chi , p , dof , expected = stats.chi2_contingency(crosstab)

if p < 0.05:
    print(Ho,"is rejected")
else:
    print(Ho,"is accepted")
```

ANOVA


Does BMI change with the no of children a woman has(one child, two children, three children)

we have to check the relation of a feature(BMI) across three categories( woman with one child, two children, three children), this is a case of ANOVA

```python
import copy
H0="No of children has no effect on bmi"

H1= "No of children has effect on bmi"

female_df = copy.deepcopy(ins[ins["sex"] == 'female'])

zero = female_df[female_df.children == 0]['bmi']
one = female_df[female_df.children == 1]['bmi']
two = female_df[female_df.children == 2]['bmi']

f_stat , p = stats.f_oneway(zero, one, two)

if p < 0.05:
    print(H0, "is rejected")
else:
    print(H0, " is accepted")
```
