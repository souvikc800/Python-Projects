## Regression Modelling

```python
action_ds = action_data[['NA_Sales', 'Global_Sales']].sort_index()
action_ds
```

```python
# Creating Feature Set
import statsmodels.api as sm

X = sm.add_constant( action_ds['NA_Sales'] )
print("NA Sales")
print(X.head(10))
```

```python
print("Global Sales")
Y = action_ds['Global_Sales']
print(Y.head(10))
```

```python
# Splitting the dataset into training and validation sets

from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split( X , Y, train_size = 0.8, random_state = 100 )
```

```python
# Datatypes

print("train_X")
print(train_X.info())
print("___________________________________________________________________")
print("test_X")
print(test_X.info())
print("___________________________________________________________________")
print("train_y",type(train_y))
print(len(train_y))
print("___________________________________________________________________")
print("test_y",type(test_y))
print(len(test_y))
```

```python
from sklearn.linear_model import LinearRegression

# Fit the Linear Regression OLS for comparison
clf =  LinearRegression()
clf.fit(train_X, train_y)
clf.score(test_X, test_y)
```

##### Inference

## Modelling Prediction

##### Making predictions in test set or new data

```python
import statsmodels.api as sm

action_lm = sm.OLS( train_y, train_X ).fit()
pred_y = action_lm.predict( test_X )
round(pred_y,2)
```

```python
from sklearn.metrics import mean_squared_error
import numpy
rmse_train = round(np.sqrt(mean_squared_error(train_y,action_lm.fittedvalues)),4)
rmse_test = round(np.sqrt(mean_squared_error(pred_y,test_y)),4)

print(" RMSE_Train: ", rmse_train)
print(" RMSE_Test: ", rmse_test)
```

```python
predictions = action_lm.get_prediction(test_X)
predictions_df = round(predictions.summary_frame(alpha=0.05),2)

# Store all the values in a dataframe
pred_y_df = pd.DataFrame( { 'NA Sales': test_X['NA_Sales'],
                            'test_y': test_y,
                            'pred_y': predictions_df['mean'],
                            'pred_y_left': predictions_df['obs_ci_lower'],
                            'pred_y_right': predictions_df['obs_ci_upper'],
                            'pred_interval_size':predictions_df['obs_ci_upper']-predictions_df['obs_ci_lower']} )
pred_y_df
```

```python
mean_interval_size = round(pred_y_df['pred_interval_size'].mean(),2)
mean_interval_size
```

##### As per our prediction model, we are getting interval size of 1.42.
